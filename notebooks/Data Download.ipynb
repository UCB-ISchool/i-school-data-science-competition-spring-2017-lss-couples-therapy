```python
name = "Brad Pitt"

def get_divorce_record(name, html_code):
    data_row = []

    data_row.append(name)
    marriage_regex = r'''title=\"married\">m\..*?[;>]\d\d\d\d'''
    divorced_regex = r'''title=\"divorced\">div\..*?[;>]\d\d\d\d'''

    marriages = re.finditer(marriage_regex, html_code)
    divorces = re.finditer(divorced_regex, html_code)

    marriage_dates = [int(marriage.group(0)[-4:]) for marriage in marriages]
    data_row.append(marriage_dates)
    data_row.append(marriage_dates[0] if len(marriage_dates) != 0 else None)

    divorce_dates = [int(divorce.group(0)[-4:]) for divorce in divorces]
    data_row.append(divorce_dates)

    marriage_lengths = []

    for divorce_index in range(len(divorce_dates)):
        marriage_lengths.append(divorce_dates[divorce_index] - marriage_dates[divorce_index])

    data_row.append(marriage_lengths)
    data_row.append(mean(marriage_lengths) if len(marriage_lengths) !=0 else None)

    return(data_row)
```

```python
def mean(numbers):
    return float(sum(numbers)) / max(len(numbers), 1)
```
